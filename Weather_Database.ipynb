```python
import pandas as pd
import numpy as np
import random
from citipy import citipy
import requests
from config import weather_api_key
```

```python
latitudes = np.random.uniform(low=-90.000, high=90.000, size=2000)
longitudes = np.random.uniform(low=-180.000, high=180.000, size=2000)
```

```python
coordinates = zip(latitudes, longitudes)
```

```python
# Create a list for holding the cities.
cities = []
countries = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = (citipy.nearest_city(coordinate[0], coordinate[1]).city_name).capitalize()
    country = (citipy.nearest_city(coordinate[0], coordinate[1]).country_code).capitalize()
    
    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
        countries.append(country)
# Print the city count to confirm sufficient count.
print(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key

citie_s = []
countrie_s = []
max_temp_l = []
min_temps = []
perc_humid_l = []
perc_cloud_l = []
wind_s = []
weather_l = []
lats = []
lons = []
i = 0


for city in cities:
    try:
        city_url = url + "&q=" + city
        city_info = requests.get(city_url)
        max_t = city_info.json()["main"]["temp_max"]
        max_temp_l.append(max_t)
        min = city_info.json()["main"]["temp_min"]
        min_temps.append(min)
        perc_h = city_info.json()["main"]["humidity"]
        perc_humid_l.append(perc_h)
        perc_c = perc_c = city_info.json()["clouds"]["all"]
        perc_cloud_l.append(perc_c)
        wind = city_info.json()["wind"]["speed"]
        wind_s.append(wind)
        weather = city_info.json()["weather"][0]["description"]
        weather_l.append(weather)
        lat = city_info.json()['coord']['lat']
        lats.append(lat)
        lon = city_info.json()['coord']['lon']
        lons.append(lon)
        cit_y = (citipy.nearest_city(lats[i], lons[i]).city_name).capitalize()
        citie_s.append(city)
        countr_y = (citipy.nearest_city(lats[i], lons[i]).country_code).capitalize()
        countrie_s.append(countr_y)
        i = i + 1
    except:
        print(f"{city} not found")

        
```

```python
Table = pd.DataFrame({
    "City": citie_s,
    "Country": countrie_s,
    "Lat": lats,
    "Lng": lons,
    "Max Temp": max_temp_l,
    "Min Temp": min_temps,    
    "Huidity": perc_humid_l,
    "Cloudiness": perc_cloud_l,
    "Wind Speed": wind_s,
    "Current Description": weather_l
})
Table.head(10)
```

```python
Table.to_csv("WeatherPy_Database.csv", index = False)
```
